#5. STANDARD DEVIATION

```python
data["fare"]
def stdev(fare):
    diffs = 0
    avg = sum(fare)/len(fare)
    for n in fare:
        diffs += (n - avg)**(2)
    return (diffs/(len(fare)-1))**(0.5)

stdev(data["fare"])
```

#1.Number of male/female VS dead/survived

```python
dmale = data[(data["sex"] == "male")  & (data["survived"] == 0)].shape[0]
dfemale = data[(data["sex"] == "female")  & (data["survived"] == 0)].shape[0]

amale = data[(data["sex"] == "male")  & (data["survived"] == 1)].shape[0]
afemale = data[(data["sex"] == "female")  & (data["survived"] == 1)].shape[0]


X = ['dead','survived']
female = [afemale,dfemale]
male = [dmale,amale]
  
X_axis = np.arange(len(X))
  
plt.bar(X_axis - 0.2, female, 0.4, label = 'Female')
plt.bar(X_axis + 0.2, male, 0.4, label = 'Male')
  
plt.xticks(X_axis, X)
plt.ylabel("Number of passengers")
plt.title("Number of males and female vs dead or survived")
plt.legend()
plt.show()
```

# Observations 

1.Number os survied pepoles are less than number dead people.

2.Maximum number of died passengers are male.
